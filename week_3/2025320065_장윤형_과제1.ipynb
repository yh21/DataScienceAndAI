```python
#todo 4 titanic 데이터로 Heat Map 시각화 하기 (8, 9번 행 완성하기)
# (조건) pivot_table, heatmap 함수를 적용할 것
# (조건) class 와 sex 필드를 적용
# (조건) blue 색상을 적용
# 8, 9번행 내용을 삭제하지 말고 추가할 것
import pandas as pd
import seaborn as sns
titanic = pd.read_excel('titanic.xlsx')
titanic_new_list = titanic.pivot_table(index='class', columns='sex', aggfunc='size')
sns.heatmap(titanic_new_list, cmap=sns.light_palette('blue', as_cmap=True), annot=True, fmt='d')
titanic_new_list
```
