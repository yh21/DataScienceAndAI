```python
#todo 3   기상청 데이터 수집 및 출력하기
# 9, 10, 11번 행 완성할 것 (기존 변수 수정, 삭제 불가)
# 다음의 2개 클래스를 탐색할 것
# ('div', {'class':'mid-tm-box upper-box'})
# ('div', {'class':'text-box'})
import requests
from bs4 import BeautifulSoup
html = requests.get('https://www.weather.go.kr/w/weather/forecast/mid-term.do?stnId1=109')
soup = BeautifulSoup(html.text, 'html.parser')
발표시각 = soup.find('div', {'class':'mid-tm-box upper-box'})
예보 = soup.find('div', {'class':'text-box'})

print(발표시각.text)
print(예보.text)
```

> 배점 2

```python
#todo 4 오늘의 탑뉴스 수집하기
# 조건1 : 7번행 괄호안에 들어갈 내용 채우기
# 조건2 : 8번행에 들어갈 함수 작성하기
import requests
from bs4 import BeautifulSoup
html = requests.get('https://news.naver.com/main/ranking/popularDay.naver')
soup = BeautifulSoup(html.content, 'html.parser')
찾은뉴스들 = soup.find_all('a', {"class": "list_title nclicks('RBP.rnknws')"})
for i in 찾은뉴스들:
    print(i.text)
```

> 배점 2


```python
#todo 5  수집된 데이터 저장하기
# 조건1: 한 행에 한줄씩 저장 되도록 6번행의 괄호 채우기
# 조건2: hotnews.txt 파일에 저장되도록 9번행 작성하기
hotnews = []
for i in 찾은뉴스들:
    hotnews.append(i.text + '\n')

with open('hotnews.txt', 'w') as f:
    f.writelines(hotnews)
# 여기에 코드 작성
```
